```python
from google import genai
from google.genai.types import Tool, GenerateContentConfig, GoogleSearch
```

```python
api_key = "<API_Key>"
```

```python


client = genai.Client(api_key=api_key)
model_id = "gemini-2.0-flash"

google_search_tool = Tool(
    google_search = GoogleSearch()
)

```

```python
from_destination = "New York"
to_destination = "Vijayawada"
budget = "INR 300000"
prompt = f"""
I want you to act as a travel advisor, you have to suggest us the the chepest flights from {from_destination} to {to_destination} and best places to visit under the selcted budget:
I need you to get the from_destination and to_destination and budget details from the user, and analyse the google data, come up with a proper
plan like hotels/hostels/dormitories to stay, some nice places to visit, recommend good food and restaurents which come in that budget, you can plan the number of days and nights, according to budget.
Plan the trip, where the user is travelling from {from_destination} to {to_destination} with a {budget}. The user is new to that place and a little importance/history/culture of that place as well.
"""
```

```python
def travel_advisor(prompt):
    response = client.models.generate_content(
    model=model_id,
    contents=[prompt],
    config=GenerateContentConfig(
        tools=[google_search_tool],
        response_modalities=["TEXT"],
    )
    )
    return response
```

```python
response = client.models.generate_content(
    model=model_id,
    contents=[prompt],
    config=GenerateContentConfig(
        tools=[google_search_tool],
        response_modalities=["TEXT"],
    )
)

for each in response.candidates[0].content.parts:
    print(each.text)
# Example response:
# The next total solar eclipse visible in the contiguous United States will be on ...

# To get grounding metadata as web content.
print(response.candidates[0].grounding_metadata.search_entry_point.rendered_content)
```
